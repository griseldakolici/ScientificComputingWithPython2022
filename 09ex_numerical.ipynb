1\. **2D minimization of a six-hump camelback function**

$$f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2$$

has multiple global and local minima.

- Find the global minima of this function
- How many global minima are there, and what is the function value at those points?
- What happens for an initial guess of $(x, y) = (0, 0)$?

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use `numpy.meshgrid()` and `pylab.imshow()` to find visually the regions.
* Use `scipy.optimize.minimize()`, optionally trying its optional arguments.

````python
#1-2D minimization of a six-hump camelback function

import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import optimize
from scipy import fftpack
from matplotlib.colors import LogNorm
from scipy.integrate import odeint
def f(x):
    return (4-2.1*x[0]**2+x[0]**4/3)*x[0]**2+ x[0]*x[1]+ (4*x[1]**2-4)*x[1]**2

x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 1, 100)


x, y = np.meshgrid(x, y)

z = np.zeros((100,100))

for i in range (100):
    for j in range (100):
        z[i,j]= f(np.array([x[i,j],y[i,j]]))


#Plot the function

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z,100)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")


minimum = optimize.minimize(f,(-0.5,-0.5)) 
#Plot the minimum found
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z,100,cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.scatter(minimum["x"][0],minimum["x"][1],f(minimum["x"]),marker="x",color = "b",s=100)


#Plot the minimum if the initial guess is (0,0)
minimum = optimize.minimize(f,(0,0))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z,100,cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.scatter(minimum["x"][0],minimum["x"][1],f(minimum["x"]),marker="x",color = "b",s=100)

````

2\. **Non-linear ODE: the damped pendulum**

The equation of the motion that a pendulum makes with respect to the angle $\theta$ with the vertical is given by:

$$\frac{d^2\theta}{dt^2} = -\frac{1}{Q} \frac{d\theta}{dt} + \sin\theta + d \cos\Omega t$$

where $t$ is time, $Q$ is the damping factor, $d$ is the forcing amplitude, and $\Omega$ is the driving frequency of the forcing. 

This second order ODE needs to be written as two coupled first order ODEs defining a new variable $\omega \equiv d\theta/dt$:

$$\frac{d\theta}{dt} = \omega$$
$$\frac{d\omega}{dt} = -\frac{1}{Q}\,\omega + \sin\theta + d \cos\Omega t$$

Consider the initial conditions $\theta_0 = \omega_0 = 0$, and $Q = 2.0$, $d = 1.5$, and $\omega = 0.65$.

 - Solve the ODE with `odeint` over a pariod of 200 time steps
 - Create two plots, one of $\theta$ as a function of the time, and $\omega$ as a function of the time
 - **Optional**: determine if there is a set of parameters for which the motion is chaotic.

````python
#2-Non-linear ODE: the damped pendulum

def pendulum(W, time):
    q=2
    d=1.5
    omega=0.65
    return(W[1], -(1/q)*W[1]+np.sin(W[0])+d*np.cos(omega*time))
W=(0,0) 
t = np.linspace(0, 100,200) 
solution = odeint(pendulum, W, t)


theta=solution[:, 0]
omega=solution[:, 1]
plt.figure(figsize=(10, 10))
plt.plot(t, theta, c="m")
plt.xlabel("time")
plt.ylabel("Theta(rad)")


plt.figure(figsize=(10, 10))
plt.plot(t, omega, c="m")
plt.xlabel("time")
plt.ylabel("Omega(rad)")
````

3\. **FFT of a simple dataset**

Perform a periodicity analysis on the lynxs-hares population, i.e. determine what is the period of the population of these animals.

The dataset is the one dowloaded at the beginning of Lecture 06:

 - `!wget https://www.dropbox.com/s/ebe1cnyd2gm836a/populations.txt -P data/`

````python
#3-FFT of a simple dataset

dataset = np.loadtxt('/Users/diogjen/Downloads/populations.txt')
year, hares, lynxes, _ = dataset.T 

plt.plot(year, lynxes, label = "Lynxes")
plt.plot(year, hares, label = "Hares")
plt.legend()
plt.show()

l_fft = fftpack.fft(lynxes)
h_fft = fftpack.fft(hares)

l_abs = np.abs(l_fft)
h_abs = np.abs(h_fft)

time_stamp = year[1]-year[0]
l_freq = fftpack.fftfreq(lynxes.size, d=time_stamp)
h_freq = fftpack.fftfreq(hares.size, d=time_stamp)

plt.plot(l_freq, l_abs, label = 'Lynxes')
plt.plot(h_freq, h_abs, label = 'Hares')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')
plt.legend()
plt.show()
````

Write a filter that removes the periodic noise from the `moonlanding.png` image by using a 2-dimensional FFT.

````python
#4-FFT of an image

#import the image as 2D numpy array
image = plt.imread("images/moonlanding.png")
plt.imshow(image)
````

````python
#Perform a 2D FFT
image_fft = fftpack.fft2(image)
image_abs = np.abs(image_fft)
plt.imshow(image_abs, norm=LogNorm(vmin=5))
plt.colorbar()
plt.show()
````

````python
#Inspect the spectrum
value = 100.
filt = np.where(image_abs <= value)
image_fft[filt] = 0
im_abs = np.abs(image_fft)
plt.imshow(im_abs, norm=LogNorm(vmin=5))
````

````python
#Apply the inverse FFT

filtered_im = fftpack.ifft2(image_fft)
plt.imshow(np.real(filtered_im))
plt.show()
````
